```python
%matplotlib inline
import pandas as pd

# Define the full file path
file_path = "/Users/JuanDiego/Documents/Proyectos Julio 2023/Customer-Churn-Data-Cleaned.csv"

df_initial_models = pd.read_csv(file_path)

```

```python
### FIRST MODEL : 
## MODEL NAME: Logistic_Regression_1
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, plot_confusion_matrix
import matplotlib.pyplot as plt


# Filter out the necessary features
features = ['Complain', 'Age', 'IsActiveMember', 'Balance', 'NumOfProducts']
X = df_initial_models[features]
y = df_initial_models['Exited']

# Splitting the data (stratified sampling)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, stratify=y, random_state=42)

# Training Logistic Regression Model
Logistic_Regression_1 = LogisticRegression()
Logistic_Regression_1.fit(X_train, y_train)

# Making predictions
y_pred = Logistic_Regression_1.predict(X_test)

# Evaluation
print("Logistic Regression Metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

```python
### IMPORTANT THOUGHTS ON WHY REMOVING COMPLAIN COULD BE BENEFITIAL TO MODEL USENESS 

###When a single feature, such as `Complain` in your dataset, exhibits such a strong correlation with the target variable, it's imperative to scrutinize its influence and potential implications.

#Here's why you might consider exploring the model without that feature:

#1. **Avoid Overfitting**: A model that overly relies on one feature can be prone to overfitting. If `Complain` is a very dominant predictor, the model may be overlooking other subtle patterns in the data that could be valuable in different scenarios.

#2. **Real-World Applicability**: Think about real-world scenarios. If the dataset was derived from a specific time frame or circumstance, there could be an overrepresentation of the relationship between `Complain` and `Exited`. By depending heavily on this feature, the model may not generalize well to new or slightly different data where the relationship might be weaker.

#3. **Model Robustness**: Relying on fewer features can sometimes lead to a more robust model. In future scenarios, if the `Complain` feature isn't available or is compromised in some way, a model dependent on it will struggle.

#4. **Understanding Other Features**: By removing the dominant feature, you might uncover other important patterns or relationships in the data that were previously overshadowed. This can lead to a more holistic understanding of the factors influencing customer churn.

#5. **Clarify Causal Relationships**: Correlation doesn't imply causation. Even if `Complain` is correlated with `Exited`, it's important to understand if it's a causal relationship. For instance, do customers exit *because* they complain, or are there underlying factors causing both the complaint and the exit? Understanding this can be crucial for business strategy.

#Given the points above, it's a good idea to test model performance without the `Complain` feature. It'll allow you to assess the importance of other features and create a model that's potentially more robust and insightful. If the model's performance drops significantly without this feature, it can confirm its dominant influence, but at least you'll have a clearer picture of the overall data landscape.
```

```python
# now i want to create 3 models with better processing techniques and without complin feature
#The idea is to find a realistic model that can adjuts to different scenarios with more common data.
## model structure as follows: Let's use feature scaling and 10 fold cross validation  with default hyperparameter tuning for each model, provide me the libraries only once, lets go one by one, use the correct model names and keep the same modeling process structure for all of them

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt


```

```python
### MODEL LOGISTIC REGRESSION
## MODEL NAME: Logistic_regression_2 (using standard scaler, Cross validation 10 fold and default features tuning)

# Features and Target
features = ['CreditScore', 'Age', 'IsActiveMember', 'Balance', 'NumOfProducts']
X = df_initial_models[features]
y = df_initial_models['Exited']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, stratify=y, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model creation and training using Logistic Regression
Logistic_Regression_2 = LogisticRegression()
Logistic_Regression_2.fit(X_train_scaled, y_train)

# Cross Validation
cv_scores = cross_val_score(Logistic_Regression_2, X_train_scaled, y_train, cv=10)
print(f"CV Average Score for Logistic Regression: {np.mean(cv_scores)}")

# Predictions
y_pred = Logistic_Regression_2.predict(X_test_scaled)

# Metrics Evaluation
print("Logistic Regression Metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_pred))


```

```python
### MODEL LOGISTIC REGRESSION
## MODEL NAME: Random_forest_2 (using standard scaler, Cross validation 10 fold and default features tuning)

# Model creation and training using Random Forest
Random_Forest_2 = RandomForestClassifier()
Random_Forest_2.fit(X_train_scaled, y_train)

# Cross Validation
cv_scores_rf = cross_val_score(Random_Forest_2, X_train_scaled, y_train, cv=10)
print(f"CV Average Score for Random Forest: {np.mean(cv_scores_rf)}")

# Predictions
y_pred_rf = Random_Forest_2.predict(X_test_scaled)

# Metrics Evaluation
print("\nRandom Forest Metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1 Score:", f1_score(y_test, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test, y_pred_rf))
```

```python
### MODEL LOGISTIC REGRESSION
## MODEL NAME: XGBoost_2 (using standard scaler, Cross validation 10 fold and default features tuning)

# Model creation and training using XGBoost
XGBoost_2 = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
XGBoost_2.fit(X_train_scaled, y_train)

# Cross Validation
cv_scores_xgb = cross_val_score(XGBoost_2, X_train_scaled, y_train, cv=10)
print(f"CV Average Score for XGBoost: {np.mean(cv_scores_xgb)}")

# Predictions
y_pred_xgb = XGBoost_2.predict(X_test_scaled)

# Metrics Evaluation
print("\nXGBoost Metrics:")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Precision:", precision_score(y_test, y_pred_xgb))
print("Recall:", recall_score(y_test, y_pred_xgb))
print("F1 Score:", f1_score(y_test, y_pred_xgb))
print("ROC AUC:", roc_auc_score(y_test, y_pred_xgb))

```

```python
#### plotting roc curve
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities
lr_probs = Logistic_Regression_2.predict_proba(X_test_scaled)[:,1]
rf_probs = Random_Forest_2.predict_proba(X_test_scaled)[:,1]
xgb_probs = XGBoost_2.predict_proba(X_test_scaled)[:,1]

# Compute ROC curve and ROC area for each model
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
lr_auc = auc(lr_fpr, lr_tpr)

rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
rf_auc = auc(rf_fpr, rf_tpr)

xgb_fpr, xgb_tpr, _ = roc_curve(y_test, xgb_probs)
xgb_auc = auc(xgb_fpr, xgb_tpr)

# Plot the ROC curves
plt.figure(figsize=(10, 7))
plt.plot(lr_fpr, lr_tpr, color='blue', linestyle='--', label=f'Logistic Regression 2 (area = {lr_auc:.2f})')
plt.plot(rf_fpr, rf_tpr, color='green', linestyle=':', label=f'Random Forest 2 (area = {rf_auc:.2f})')
plt.plot(xgb_fpr, xgb_tpr, color='red', linestyle='-.', label=f'XGBoost 2 (area = {xgb_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
